```python
import cv2
import numpy as np
import os 
from random import shuffle
```

```python
train_directory='Downloads/dogs-vs-cats/train'
test_directory='Downloads/dogs-vs-cats/test1'
img_size=50
lr=1e-3
model_name='cats-dogs-{}-{}.model'.format(lr,'conv-basic')
```

```python
def create_train_data():
    train_data=[]
    
    for img in os.listdir(train_directory):
        label= label_f(img) #write a function which returns 1 or 0
        
        path=os.path.join(train_directory,img)
        
        train_sample=cv2.resize(cv2.imread(path,cv2.IMREAD_GRAYSCALE),(img_size,img_size))
        
        train_data.append([np.array(train_sample),(label)])
        
    shuffle(train_data)
    np.save('train_data.npy',train_data)
    return train_data
```

```python
def label_f(img):
    
    word_label=img.split('.')[-3]
    if word_label== "dog": return 1
    elif word_label == "cat": return 0
```

```python
def create_submission_data():
    sub_data=[]
    
    for img in os.listdir(test_directory):
        
        img_num=img.split('.')[0]
        
        path=os.path.join(test_directory,img)
        
        sub_sample=cv2.resize(cv2.imread(path,cv2.IMREAD_GRAYSCALE),(img_size,img_size))
        
        sub_data.append([np.array(sub_sample),np.array(img_num)])
        
    np.save('submission_data.npy',sub_data)
    
    return sub_data
```

```python
X=create_train_data()
```

```python
train_datas=X[:-500]
test_datas=X[-500:]
```

```python
train_x=np.array([i[0] for i in train_datas]).reshape(-1, img_size, img_size,1)
train_y=[i[1] for i in train_datas]


test_x=np.array([i[0] for i in test_datas]).reshape(-1, img_size, img_size,1)
test_y=[i[1] for i in test_datas]


train_x,test_x=train_x/255.0,test_x/255.0
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Activation, Flatten,Dense
```

```python
model=Sequential()

model.add(Conv2D(64,(5,5),input_shape=train_x.shape[1:]))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64,(3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))


model.add(Conv2D(64,(3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())

model.add(Dense(64))

model.add(Dense(1))
model.add(Activation("sigmoid"))

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

model.fit(train_x,train_y,batch_size=32,validation_split=0.1,epochs=3)



```

```python
model.predict(train_x)
```

```python
model.evaluate(test_x, test_y,  verbose=1)
```

```python
model.metrics_names
```

```python
submission_data=create_submission_data()
sub_data=np.array([i[0] for i in submission_data]).reshape(-1, img_size, img_size,1)
```

```python
sub_data=sub_data/255.0
sub_pred=model.predict(sub_data)
print(sub_pred)
```
